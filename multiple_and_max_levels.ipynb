```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "/home/hp/Documents/BRSM/data_brsm/Combined.csv"
df = pd.read_csv(file_path)
df_filtered = df[df["target_load"] == "Multiple"]
players = list(range(22, 38))
df_filtered = df_filtered[df_filtered["participant"].isin(players)]

player_avg = df_filtered.groupby(["participant","modality"])[[
    "RT_mean(ms)",
    "Initial_RT(ms)",
    "Accuracy_mean"
]].mean().reset_index()

sns.set(style="whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(20,6))

sns.barplot(data=player_avg, x="participant", y="RT_mean(ms)", hue="modality", ax=axes[0])
axes[0].set_title("RT_mean")
axes[0].set_xlabel("Player")
axes[0].set_ylabel("RT_mean (ms)")

sns.barplot(data=player_avg, x="participant", y="Initial_RT(ms)", hue="modality", ax=axes[1])
axes[1].set_title("Initial RT")
axes[1].set_xlabel("Player")
axes[1].set_ylabel("Initial RT (ms)")

sns.barplot(data=player_avg, x="participant", y="Accuracy_mean", hue="modality", ax=axes[2])
axes[2].set_title("Accuracy")
axes[2].set_xlabel("Player")
axes[2].set_ylabel("Accuracy")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "/home/hp/Documents/BRSM/data_brsm/Combined.csv"
df = pd.read_csv(file_path)

df = df[df["target_load"] == "Multiple"]

players = list(range(22, 38))
df = df[df["participant"].isin(players)]

df.loc[df["Accuracy_mean"] > 1, "Accuracy_mean"] = df["Accuracy_mean"] / 100

sns.set(style="whitegrid")

fig, axes = plt.subplots(2, 3, figsize=(18,10))

sns.violinplot(data=df, x="modality", y="RT_mean(ms)", ax=axes[0,0])
sns.boxplot(data=df, x="modality", y="RT_mean(ms)", ax=axes[1,0])

sns.violinplot(data=df, x="modality", y="Initial_RT(ms)", ax=axes[0,1])
sns.boxplot(data=df, x="modality", y="Initial_RT(ms)", ax=axes[1,1])

sns.violinplot(data=df, x="modality", y="Accuracy_mean", ax=axes[0,2])
sns.boxplot(data=df, x="modality", y="Accuracy_mean", ax=axes[1,2])

axes[0,0].set_title("RT_mean Violin")
axes[1,0].set_title("RT_mean Box")

axes[0,1].set_title("Initial_RT Violin")
axes[1,1].set_title("Initial_RT Box")

axes[0,2].set_title("Accuracy Violin")
axes[1,2].set_title("Accuracy Box")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Combined.csv")
df = df[df["target_load"] == "Single"]
players = list(range(1, 22))
df = df[df["participant"].isin(players)]
df.loc[df["Accuracy_mean"] > 1, "Accuracy_mean"] = df["Accuracy_mean"] / 100

player_avg = df.groupby(["participant","modality"])[[
    "RT_mean(ms)",
    "Accuracy_mean"
]].mean().reset_index()

sns.set(style="whitegrid")
fig, axes = plt.subplots(1, 2, figsize=(18,6))

sns.barplot(
    data=player_avg,
    x="participant",
    y="RT_mean(ms)",
    hue="modality",
    ax=axes[0]
)
axes[0].set_title("RT_mean Comparison (Players 1–21)")
axes[0].set_xlabel("Participant")
axes[0].set_ylabel("RT_mean (ms)")
sns.barplot(
    data=player_avg,
    x="participant",
    y="Accuracy_mean",
    hue="modality",
    ax=axes[1]
)
axes[1].set_title("Accuracy Comparison (Players 1–21)")
axes[1].set_xlabel("Participant")
axes[1].set_ylabel("Accuracy")
axes[1].legend_.remove()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

single = pd.read_csv("/home/hp/Documents/BRSM/data_brsm/player_max_levelsingle.csv")
multiple = pd.read_csv("/home/hp/Documents/BRSM/data_brsm/player_max_levels.csv")

single = single.rename(columns={"MaxLevelReached": "SinglePhone"})
multiple = multiple.rename(columns={"MaxLevelReached": "MultiplePhone"})

single["PlayerID"] = single["PlayerID"].astype(str).str.extract(r'(\d+)').astype(int)
multiple["PlayerID"] = multiple["PlayerID"].astype(str).str.extract(r'(\d+)').astype(int)

df = pd.merge(single, multiple, on="PlayerID", how="outer")

df = df.sort_values("PlayerID")

print(df)

plt.figure(figsize=(10,6))

plt.plot(df["PlayerID"], df["SinglePhone"], marker='o', linewidth=2, label="Single Phone")
plt.plot(df["PlayerID"], df["MultiplePhone"], marker='s', linewidth=2, label="Multiple Phone")

plt.xlabel("Player ID")
plt.ylabel("Maximum Level Reached")
plt.title("Maximum Level Reached by Each Player (Phone)")

plt.xticks(df["PlayerID"])
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

single = pd.read_csv("/home/hp/Documents/BRSM/data_brsm/player_max_levelsingle.csv")
multiple = pd.read_csv("/home/hp/Documents/BRSM/data_brsm/player_max_levels.csv")

single["PlayerID"] = single["PlayerID"].astype(str).str.extract(r'(\d+)').astype(int)
multiple["PlayerID"] = multiple["PlayerID"].astype(str).str.extract(r'(\d+)').astype(int)

single_freq = single["MaxLevelReached"].value_counts().sort_index()
multiple_freq = multiple["MaxLevelReached"].value_counts().sort_index()

freq_df = pd.DataFrame({
    "SinglePhone": single_freq,
    "MultiplePhone": multiple_freq
}).fillna(0)

print(freq_df)

freq_df.plot(kind="bar", figsize=(10,6))

plt.xlabel("Maximum Level Reached")
plt.ylabel("Number of Players")
plt.title("Frequency of Maximum Levels Reached (Phone)")
plt.grid(axis="y")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

combined = pd.read_csv("Combined.csv")
order_single = pd.read_csv("player_start_comparison_single.csv")

combined["participant"] = combined["participant"].astype(int)

order_single["participant"] = (
    order_single["participant"]
    .astype(str)
    .str.extract("(\d+)")[0]
    .astype(int)
)

combined["modality"] = combined["modality"].str.lower()
combined["target_load"] = combined["target_load"].str.lower()

single = combined[combined["target_load"] == "single"]

data = pd.merge(
    single,
    order_single[["participant","started_first"]],
    on="participant",
    how="inner"
)

data = data.sort_values("participant")

phone = data[data["modality"] == "phone"]
lab = data[data["modality"] == "lab"]

plt.figure(figsize=(10,6))

plt.plot(phone["participant"], phone["RT_mean(ms)"], color="blue", linewidth=2, label="Phone RT")
plt.plot(lab["participant"], lab["RT_mean(ms)"], color="orange", linewidth=2, label="Lab RT")

order_y = max(data["RT_mean(ms)"]) + 300

plt.plot(data["participant"], [order_y]*len(data["participant"]), color="gray", linewidth=1)

for _, row in data.iterrows():

    marker = "o" if row["started_first"] == "phone" else "s"
    color = "blue" if row["started_first"] == "phone" else "orange"

    plt.scatter(
        row["participant"],
        order_y,
        marker=marker,
        color=color,
        s=120
    )

plt.xticks(range(1,22))

plt.title("Single Target: RT_mean Comparison (Phone vs Lab)")
plt.xlabel("Participant Number")
plt.ylabel("RT_mean (ms)")

legend_elements = [
    Line2D([0], [0], color='blue', lw=2, label='Phone RT'),
    Line2D([0], [0], color='orange', lw=2, label='Lab RT'),
    Line2D([0], [0], marker='o', color='blue', linestyle='None', markersize=8, label='Phone Played First'),
    Line2D([0], [0], marker='s', color='orange', linestyle='None', markersize=8, label='Lab Played First')
]

plt.legend(handles=legend_elements)

plt.tight_layout()
plt.show()
```
